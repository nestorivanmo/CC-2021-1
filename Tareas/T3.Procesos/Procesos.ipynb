# Programa 1
Realiza el programa que instancie, como clase o con el método ```Process```, 10 procesos. Cada uno de los procesos hijos recibirán un valor entero y un caracter $(i, c)$ enviados por el proceso padre, los procesos escribirán en la salida estándar $i$ veces el caracter $c$.

````python
import multiprocessing as mp
import numpy as np
import string, random, time, sys, os
````

````python
def escribe(i, c, num_proceso):
	"""Imprime i veces el número del proceso, identificador del sistema operativo y el caracter.

	Argumentos:
	i -- número de veces a imprimir el caracter c
	c -- caracter a imprimir
	num_proceso -- número del proceso
	"""
	for n in range(i):
		sys.stdout.write(f"Soy el proceso #{num_proceso} con id {os.getpid()}, mi caracter es: {c}\n")


if __name__ == '__main__':
	procesos = [0]*10
	for i in range(10):
		num_times = np.random.randint(1,10)
		proc_char = random.choice(string.ascii_letters)
		proc_id = i+1
		procesos[i] = mp.Process(target=escribe, args=(num_times, proc_char, proc_id))
		procesos[i].start()
	for i in range(10):
		procesos[i].join()
````

# Programa 2
Refactoriza (reescribe) el programa anterior y elabora el **Programa 2** que incluya un mecanismo de sincronización el cual permita escribir en orden todos los caracteres de cada proceso. Es decir, que se obtenga la secuencia $c_{1,1},\dots,c_{1,i}, c_{2,1},\dots, c_{2,i}, \dots, c_{10,1},\dots , c_{10,i}$ donde cada subsecuencia $c_{k,i}$para cada $k= 1,2, ...,10$ es la secuencia de caracteres del proceso hijo $k$ con longitud $i$.

````python
def escribe(i, c, num_proceso, turno):
	"""Imprime i veces el caracter c siempre y cuando sea el turno del proceso

	Argumentos:
	i -- número de veces a imprimir el caracter c
	c -- caracter a imprimir
	num_proceso -- número del proceso
	turno   	-- multiprocessing.Value cuyo atributo value indica el turno actual
	"""
	while turno.value != num_proceso: # espera a que sea su turno
		continue
	sys.stdout.write(f"(i:{i}, c:\'{c}\')\n")
	sys.stdout.write(f"Soy el proceso #{num_proceso} con id {os.getpid()}:\n")
	for n in range(i):
		sys.stdout.write(f"{c}\n")
	sys.stdout.write("-------------------------------\n")
	turno.value += 1

if __name__ == '__main__':
	procesos = [0]*10
	turno = mp.Value('i', 1) # indica el turno actual
	for i in range(10):
		num_times = np.random.randint(1,10)
		proc_char = random.choice(string.ascii_letters)
		proc_id = i+1
		procesos[i] = mp.Process(target=escribe, args=(num_times, proc_char, proc_id, turno))
		procesos[i].start()
	for i in range(10):
		procesos[i].join()
````

# Programa 3
Refactoriza (reescribe) el **Programa 2** y elabora el **Programa 3** donde construyas un mecanismo de sincronización el cual permita escribir en orden todos los caracteres de cada proceso siguiendo una política de orden $p$ que será una lista de números enteros aleatorios con los índices $k$ de cada proceso hijo. La escritura de los caracteres seguirá la secuencia de $c_{k,i}(p)$donde cada secuencia $c_{k,i}$ estará definida por la política $p$.

````python
def escribe(i, c, num_proceso, orden_de_ejecucion, indice_orden):
	"""Imprime i veces el caracter c siempre y cuando el num_proceso corresponda
	con el orden de ejecución definido previamente.

	Argumentos:
	i                   -- número de veces a imprimir el caracter c
	c                   -- caracter a imprimir
	num_proceso         -- número del proceso
	orden_de_ejecucion  -- lista con enteros (en nuestro caso, del 1 al 10)
	indice_orden        -- índice que, al combinarse con orden_de_ejecucion
							 indican el número del proceso a ejecutar
	"""
	while orden_de_ejecucion[indice_orden.value] != num_proceso:
		continue
	sys.stdout.write(f"(i:{i}, c:\'{c}\')\n")
	sys.stdout.write(f"Soy el proceso #{num_proceso} con id {os.getpid()}:\n")
	for n in range(i):
		sys.stdout.write(f"{c}\n")
	sys.stdout.write("--------------------------------\n")
	indice_orden.value += 1

if __name__ == '__main__':
	procesos = [0]*10
	orden_de_ejecucion = list(range(1, 11))
	random.shuffle(orden_de_ejecucion) # este será el orden de ejecución de los procesos
	sys.stdout.write(f"El orden de ejecución será: {orden_de_ejecucion}\n\n")
	indice_orden = mp.Value('i', 0) # es un indice que indica el proceso que debe ejecutarse
	for i in range(10):
		num_times = np.random.randint(1,10)
		proc_char = random.choice(string.ascii_letters)
		proc_id = i+1
		procesos[i] = mp.Process(
			target=escribe,
			args = (num_times, proc_char, proc_id, orden_de_ejecucion, indice_orden)
		)
		procesos[i].start()
	for i in range(10):
		procesos[i].join()
````
